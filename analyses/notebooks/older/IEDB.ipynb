```python
def build_variant(row, genome):
    from varcode import Variant
    return Variant(
        contig=row["chr"],
        ref=row["ref"],
        alt=row["alt"],
        start=row["start"],
        ensembl=genome)
```

```python
def wildtype_peptide(row, genome, effects):
    variant = build_variant(row, genome)
    effects = effects.filter(lambda effect: effect.variant == variant)
    assert len(effects) == 1, "Expected a single effect to match the variant %s" % variant
    effect = effects[0]
    peptide_start_in_protein = row["peptide_start_in_protein"]
    peptide = row["peptide"]
    length = row["length"]
    assert effect.mutant_protein_sequence[
        peptide_start_in_protein:peptide_start_in_protein + length] == peptide, (
            "Mutant protein sequence slice should equal peptide")
    return effect.original_protein_sequence[
        peptide_start_in_protein:peptide_start_in_protein + length] 
```

```python
def grab_wt_sequence(row):
    effect = row['effect'].upper().split(".")[1]
    assert '*' not in effect
    effect_location = int(''.join([el for el in effect if not el.isalpha()])) - 1
    effect_to_from = [el for el in effect if el.isalpha()]
    assert effect_location == row['protein_mutation_start'], "Effect location %s doesn't match %s" % (effect_location, row['protein_mutation_start'])
    assert len(effect_to_from) == 2, "Error: %s / %s" % (effect_to_from, effect)
    effect_to, effect_from = effect_to_from
    seq = row['protein_sequence']
    assert seq[effect_location] == effect_from
    seq = seq[:effect_location] + effect_to + seq[(effect_location + 1):]
    start = int(row['peptide_offset_in_protein'])
    end = start + int(row['length'])
    wt_seq = seq[start:end]
    assert wt_seq != row['peptide'], "Same sequence: %s (wt) and %s" % (wt_seq, row['peptide'])
    return seq[start:end]
```
