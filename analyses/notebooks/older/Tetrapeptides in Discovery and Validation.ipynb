```python
from utils.tetrapeptides import get_kmers
```

```python
def index_epitopes(iedb_df, column):
    from collections import defaultdict
    indexed_epitopes = defaultdict(list)
    for _, row in iedb_df.iterrows():
        seq = row[column]
        # Get kmers from IEDB epitopes
        substrings = get_kmers(seq, kmer_size=4)
        for sub in substrings:
            indexed_epitopes[sub] += [row]
    return indexed_epitopes
```
